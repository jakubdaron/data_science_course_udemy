Obliczanie nowych kolumn

```python
df = fetch_finacial_data('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head(3)
```

```python
df['Average'] = (df.Open + df.Close) / 2.0
df.head()
```

```python
df = df.sort_index()
df.head()
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
df['Daily_change'] = (df.Close / df.Close_shift - 1) * 100
df.head()
```

```python
del df['Daily_change[%]']
```

```python
df.head()
```
